# Q2. gasa를 띄어쓰기 기준으로 tokenize하고, index의 sequence로 바꾸어 내시오.

**조건**
0. gasa_seq에, index의 seq로 바뀐 가사를 선언해둔다.
1. 데이터가 부족하다, 모든 단어 다 사용할 것.

````python
from tensorflow.keras.preprocessing.text import Tokenizer
````

````python
####################
## Your Code here ##
####################
tokenizer = Tokenizer()
tokenizer.fit_on_texts(gasa)
# 0. gasa_seq에, index의 seq로 바뀐 가사를 선언해둔다.
gasa_seq = tokenizer.texts_to_sequences(gasa)
# 1. 데이터가 부족하다, 모든 단어 다 사용할 것.
# 2. vocab_size라는 변수에 사용하게 될 token의 수를 선언해두자.
vocab_size = len(tokenizer.index_word) + 1

print(vocab_size)
````

# Q3. gasa_seq안의 모든 문장을 다음과 같이 재구성하시오.

**예시**
* gasa_seq가 [ [1,2,3], [7,2,9,10] ] 라면
* 결과 : gasa_seq = [ [1, 2], [1,2,3], [7,2], [7,2,9], [7,2,9,10] ]
* 중간에 다른 변수를 임시로 사용해도 좋다.

````python
####################
## Your Code here ##
####################
array = []
for seq in gasa_seq :
    for i in range(2, len(seq)) :
        array.append(seq[:i])

````

# Q4. gasa_seq안의 가장 긴 문장을 찾고, 그 길이에 맞추어 문장 길이를 통일하시오.

**Questions**
1. 0을 앞에 덧대야 할까, 뒤에 덧대야 할까?
2. 문장의 길이가 매우 길다면, 가장 긴 문장에 맞추어 길이를 통일하는 것이 좋은 방법일까?
3. gasa_seq의 shape를 확인하라.

````python
####################
## Your Code here ##
####################
from tensorflow.keras.preprocessing.sequence import pad_sequences #문장길이통일기

max_len = max( [ len(seq) for seq in array]      )
# gasa_seq 리스트 안에있는 모든 시퀀스(seq)들의 길이를 측정하는 리스트를 만들고
# 그 리스트의 가장 최대값을 뽑으면
# 그게 가장 긴 시퀀스의 길이 아닌가!
gasa_seq = pad_sequences(array, maxlen=max_len, padding='pre')
````

# Q5. gasa_seq의 마지막 컬럼은 y로, 나머지 컬럼은 x로 분할하라.

````python
####################
## Your Code here ##
####################
# x와 y를 미리 쪼개놓고 코드 짜면 귀찮으니까
# 한번에 처리 한다음
# 마지막에 쪼개기
# 여기를 이해한다음 다시 Q3을 감상하시오.
x = gasa_seq[:, : -1] 
y = gasa_seq[:, -1]

````

# Q6. y를, vocab_size에 맞추어 원핫 인코딩 하시오.

1. num_classes 를 반드시 vocab_size에 맞출 것.
2. 사용하지 않을 컬럼이 존재할 수 있다.
    * 예시) y에는 4가 없는데, 0으로만 가득찬 4번 컬럼이 만들어질 수도 있다.

**추가설명**
* 효율을 위해, 이 모든걸 엄격히 보정하여 사용할 수도 있다.
* 그러나, 코드가 매우 길어진다.
* 데이터가 많아지면, 별차이도 없어진다.

````python
####################
## Your Code here ##
####################
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes = vocab_size)

````

# Q7. 문장을 받아, 다음 단어를 선택하는 모델을 만드시오.

* 컴파일까지 마치시오.
    * 모니터링 지표로 accuracy 사용0
* 모델 구조는 아래와 같다.
    1. 임베딩 레이어 : 임베딩 차원 32, input_length 주의. gasa_seq를 x와 y로 분할했기 때문에, 문장길이가 달라졌다.
    2. GRU, 32, 모든 출력을 다음 layer로
    3. GRU, 48, 모든 출력을 다음 layer로
    4. GRU, 64, 마지막 출력만 다음 레이어로.
    5. 멀티클래스 분류 아웃풋레이어


````python
####################
## Your Code here ##
####################
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import GRU, Embedding, Dense

````

````python
####################
## Your Code here ##
####################
keras.backend.clear_session()

model = keras.models.Sequential()

model.add( keras.layers.Embedding(input_dim = vocab_size,
                                  output_dim = 32,
                                  input_length = max_len-1 ))
model.add( GRU(32, return_sequences=True) )
model.add( GRU(48, return_sequences=True) )
model.add( GRU(64, return_sequences=False) )
model.add( Dense( vocab_size, activation='softmax' ))

model.compile( loss = 'categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()

````

# Q8. 모델을 학습시킬 것.

* batch_size = 64
* epochs = 1000


````python
####################
## Your Code here ##
####################

model.fit(x, y, epochs=1000, batch_size=64, verbose=1)
````

# Q11. 9번,10번 문항을 참고하여, 다음의 기능을 하는 함수를 제작한다.

* 편의를 위해, 외부에 선언된 변수를 활용해도 좋다.
* 인풋 : 사용할 첫 단어, 반복 예측 수
* 내부 : 매번 가장 확률 높은 단어만 선택한다.
* 아웃풋 : 완성된 문장.
* 예시 :
```
jaksa("나", 4 )
>> "나 추억 한다면 힘차게 걸으리라"
```
* 실제로 노래 가사에 사용된 여러 단어를 넣어 테스트해보자.

````python
####################
## Your Code here ##
####################
def jaksa( word, n_iter):
    for i in range( n_iter ) :
        words = [word]
        seq_ = tokenizer.texts_to_sequences(words)
        seq_ = pad_sequences(seq_, maxlen=max_len-1, padding='pre')
        pred_ = model.predict(seq_, )
        top3_idx = np.argsort(pred_)[:,-3:].reshape(-1)
        # top3_probs = pred_[:, top3_idx].reshape(-1) # 군더더기
        next_word_idx = top3_idx[-1]
        next_word = tokenizer.index_word[ next_word_idx ]

        word = word + " " + next_word
    return word

````
