```python
import tensorflow as ft
from tensorflow import keras
```

```python
## Functional API

# 1번 세션 클리어
keras.backend.clear_session()
# 2번 레이어 사슬처럼 엮기
il = keras.layers.Input(shape = (5,5,3))
cl = keras.layers.Conv2D(filters = 3,
                         kernel_size = (3,3),
                         strides = (1,1),
                         activation = 'relu')(il)
pl = keras.layers.MaxPool2D(pool_size=(2,2),
                            )(cl)
# 3번 모델의 시작과 끝 지정
model = keras.models.Model(il, pl)

# summary()
model.summary()
```
